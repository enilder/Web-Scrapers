```python
import requests as req
from bs4 import BeautifulSoup
import re
```

```python
url = "http://www.imdb.com/title/tt4425200/ratings?ref_=tt_ov_rt"
```

```python
review_page = req.get(url, headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.153 Safari/537.36'})
```

```python
html = BeautifulSoup(review_page.text, "html.parser")
```
